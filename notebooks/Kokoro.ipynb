```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import torch
```

```python
pipeline = KPipeline(lang_code='a')

text = '''
Shows the specific function equation based on the current a, b, and c values. In the example snapshot, the equation shown is f(x) = (x + 2) / (x² - 3.3x + 2). (Note: the slider for 'b' shows -2.4, while the equation uses -3.3, suggesting the snapshot might have been taken mid-adjustment or there's a slight mismatch in the display). Despite its lightweight architecture, it delivers comparable quality to larger models while being significantly faster and more cost-efficient.
'''
```

```python
generator = pipeline(text, voice='af_heart')
for i, (gs, ps, audio) in enumerate(generator):
    print(i, gs, ps)
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    sf.write(f'{i}.wav', audio, 24000)
```
