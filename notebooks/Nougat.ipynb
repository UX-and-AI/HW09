```python
from huggingface_hub import hf_hub_download
import re
from PIL import Image
#https://huggingface.co/docs/transformers/main/en/model_doc/nougat#transformers.NougatProcessor
```

```python
from transformers import NougatProcessor, VisionEncoderDecoderModel
from datasets import load_dataset
import torch
```

```python
processor = NougatProcessor.from_pretrained("facebook/nougat-base")
model = VisionEncoderDecoderModel.from_pretrained("facebook/nougat-base")
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
# prepare PDF image for the model
# filepath = hf_hub_download(repo_id="hf-internal-testing/fixtures_docvqa", filename="Mid-Term Presentation _page-0006.jpg", repo_type="dataset")
# image = Image.open(filepath)
# pixel_values = processor(image, return_tensors="pt").pixel_values
```

```python

image = Image.open("Tests/MathTest2.png").convert("RGB") 
pixel_values = processor(image, return_tensors="pt").pixel_values
```

```python
outputs = model.generate(
    pixel_values.to(device),
    min_length=1,
    max_new_tokens=2000,
    bad_words_ids=[[processor.tokenizer.unk_token_id]],
)
```

```python
sequence = processor.batch_decode(outputs, skip_special_tokens=True)[0]
sequence = processor.post_process_generation(sequence, fix_markdown=False)
```
